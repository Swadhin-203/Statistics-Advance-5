```python
# Q1. Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation 
# of 5 using Python. Interpret the results.
from scipy import stats

# Set up the sample data
sample_mean = 50
sample_std = 5
sample_size = 100

# Calculate the standard error
standard_error = sample_std / (sample_size ** 0.5)

# Calculate the 95% confidence interval
conf_int = stats.norm.interval(0.95, loc=sample_mean, scale=standard_error)

print("The 95% confidence interval is:", conf_int)
```

```python
# Q4. A study of the prevalence of smoking in a population of 500 individuals found that 60 individuals
# smoked. Use Python to calculate the 95% confidence interval for the true proportion of individuals in the
# population who smoke.
# Ans:
from statsmodels.stats.proportion import proportion_confint

# Define the sample size and number of individuals who smoke
n = 500
x = 60

# Calculate the 95% confidence interval for the true proportion
conf_int = proportion_confint(count=x, nobs=n, alpha=0.05, method='wilson')

print(f"95% Confidence interval: {conf_int}")
```

```python
# Q5. Calculate the 90% confidence interval for a sample of data with a mean of 75 and a standard deviation
# of 12 using Python. Interpret the results.
# Ans
from scipy.stats import norm

# Define the sample mean and standard deviation
mean = 75
std_dev = 12

# Define the sample size
n = 100

# Calculate the 90% confidence interval
conf_int = norm.interval(alpha=0.90, loc=mean, scale=std_dev / (n ** 0.5))

print(f"90% Confidence interval: {conf_int}")
```

```python
# Q6. Use Python to plot the chi-square distribution with 10 degrees of freedom. Label the axes and shade the
# area corresponding to a chi-square statistic of 15.
# Ans
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Define the degrees of freedom
df = 10

# Define the x values to plot
x = np.linspace(0, 30, 500)

# Define the chi-square distribution with 10 degrees of freedom
chi2 = stats.chi2(df)

# Calculate the y values for the chi-square distribution
y = chi2.pdf(x)

# Plot the chi-square distribution
plt.plot(x, y)

# Shade the area corresponding to a chi-square statistic of 15
plt.fill_between(x[x >= 15], y[x >= 15], color='orange')

# Add labels and title to the plot
plt.xlabel('Chi-square statistic')
plt.ylabel('Probability density')
plt.title('Chi-square distribution with 10 degrees of freedom')

# Show the plot
plt.show()
```

```python
# Q7. A random sample of 1000 people was asked if they preferred Coke or Pepsi. Of the sample, 520
# preferred Coke. Calculate a 99% confidence interval for the true proportion of people in the population who
# prefer Coke.
# Ans:
from scipy.stats import norm

# Define the sample proportion
sample_proportion = 520 / 1000

# Define the sample size
n = 1000

# Calculate the z-score associated with the 99% confidence level
z = norm.ppf(0.995)

# Calculate the confidence interval
lower = sample_proportion - z * ((sample_proportion * (1 - sample_proportion)) / n) ** 0.5
upper = sample_proportion + z * ((sample_proportion * (1 - sample_proportion)) / n) ** 0.5

# Print the confidence interval
print(f"99% Confidence interval: ({lower:.4f}, {upper:.4f})")
```

```python
# Q8. A researcher hypothesizes that a coin is biased towards tails. They flip the coin 100 times and observe
# 45 tails. Conduct a chi-square goodness of fit test to determine if the observed frequencies match the
# expected frequencies of a fair coin. Use a significance level of 0.05.
# Ans
from scipy.stats import chisquare

# Define the observed and expected frequencies
observed = [55, 45]  # 55 heads, 45 tails
expected = [50, 50]  # equal probability of heads and tails

# Calculate the chi-square statistic and p-value
chi2, p = chisquare(observed, expected)

# Print the results
print(f"Chi-square statistic: {chi2:.4f}")
print(f"P-value: {p:.4f}")

# The chi-square statistic is 1.0000, and the p-value is 0.3173. At a significance level of 0.05, the critical value for the
# chi-square distribution with 1 degree of freedom is 3.84. Since our calculated chi-square statistic (1.0000) is less than the 
# critical value (3.84), we fail to reject the null hypothesis that the observed frequencies match the expected frequencies for 
# a fair coin. In other words, we do not have sufficient evidence to conclude that the coin is biased towards tails.
```

# Q11. A random sample of 30 people was selected from a population with an unknown mean and standard deviation. The sample mean was found to be 72 and the sample standard deviation was found to be 10. Conduct a hypothesis test to determine if the population mean is significantly different from 70. Use a significance level of 0.05.
Ans:To conduct a hypothesis test for the population mean, we can use a t-test since the population standard deviation is unknown and the sample size is less than 30.

The null and alternative hypotheses are:

H0:  = 70 (the population mean is equal to 70)
Ha:   70 (the population mean is not equal to 70)
We can use the t-distribution with n-1 degrees of freedom to calculate the p-value for the test statistic. We will use a two-tailed test with a significance level of 0.05.

Here's the Python code to conduct the hypothesis test:

```python
from scipy.stats import t

# define the sample size, sample mean, and sample standard deviation
n = 30
x_bar = 72
s = 10

# define the null hypothesis and hypothesized population mean
mu0 = 70

# calculate the test statistic
t_stat = (x_bar - mu0) / (s / (n ** 0.5))

# calculate the p-value
p_val = 2 * (1 - t.cdf(abs(t_stat), df=n-1))

# print the results
print("Test statistic:", t_stat)
print("p-value:", p_val)
```
